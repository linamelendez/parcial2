```
numeros = []
for i in range(1, 6):
    while True:
        entrada = input(f"Ingresa el número {i}: ")
        if entrada == "":
            print("  No puedes dejar el campo vacío. Intenta de nuevo.")
        else:
            numero = float(entrada)
            numeros.append(numero)
            break

mayor = numeros[0]
menor = numeros[0]
suma = 0

for n in numeros:
    if n > mayor:
        mayor = n
    if n < menor:
        menor = n
    suma = suma + n

promedio = suma / 5
print("Número mayor:", mayor)
print("Número menor:", menor)
print("Promedio:", promedio)
```

```
def es_primo(numero):
    if numero < 2:
        return False

    divisores = 0

    for i in range(1, numero + 1):
        if numero % i == 0:
            divisores = divisores + 1

    if divisores == 2:
        return True
    else:
        return False
num = int(input("Ingresa un número: "))

if es_primo(num):
    print(num, "ES un número primo")
else:
    print(num, "NO es un número primo")
```

```
numeros = [3, 7, 12, 18, 21, 30]
pares = []

for numero in numeros:
    if numero % 2 == 0:
        pares.append(numero)

print("Lista original:", numeros)
print("Lista de pares:", pares)
```

```
def invertir_palabra(palabra):
    invertida = ""

    for letra in range(len(palabra) - 1, -1, -1):
        invertida = invertida + palabra[letra]

    return invertida
palabra = input("Ingresa una palabra: ")
resultado = invertir_palabra(palabra)
print("Palabra original: ", palabra)
print("Palabra invertida:", resultado)
```

```
def contarLetra(palabra, letra):
    contador = 0

    for caracter in palabra:
        if caracter == letra:
            contador = contador + 1

    return contador

palabra = input("Ingresa una palabra: ")
letra = input("Ingresa la letra a buscar: ")

resultado = contarLetra(palabra, letra)

print("La letra", letra, "aparece", resultado, "veces en la palabra", palabra)
```

```
def factorial(numero):
    if numero == 0:
        return 1

    resultado = 1

    for i in range(1, numero + 1):
        resultado = resultado * i

    return resultado

num = int(input("Ingresa un número: "))
print("El factorial de", num, "es:", factorial(num))
```
